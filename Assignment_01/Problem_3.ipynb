```python
from scipy.stats import norm
import numpy as np
import matplotlib.pyplot as plt
```

```python
loc, scale = 0., 1.
s = np.random.laplace(loc, scale, 1000)
x = np.arange(-4., 4., .01)
pdf = np.exp(-abs(x-loc)/scale)/(2.*scale)
plt.plot(x, norm.pdf(x, 0, 1))
plt.plot(x, pdf)
plt.legend(['p(x|η)','p(x|γ)'])
plt.xlabel("x")
plt.ylabel("Density")
```
